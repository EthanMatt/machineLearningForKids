## 1. 什么是机器学习？想象一下：你是一个天气预报员。**传统方式**：科学家告诉你规则——"湿度超过80%就容易下雨"**机器学习方式**：你看了1000天的天气数据，自己总结出规则机器学习就是让 **计算机从数据中自己学习规律**，而不是人类手动写规则。

### 机器学习的两种类型**监督学习 (Supervised Learning)**：给计算机看"题目+答案"，让它学会举一反三- 比如：给它看1000天的天气和是否下雨，让它预测明天**无监督学习 (Unsupervised Learning)**：只给计算机看"题目"，让它自己发现规律- 比如：给它看很多城市的天气，让它自己把相似的城市分成一组

## 2. 数据可视化：用眼睛发现规律俗话说"一图胜千言"。画图是机器学习中 **最重要** 的技能之一！

```python
# 画一个散点图：温度 vs 湿度plt.figure(figsize=(10, 6))# 按城市用不同颜色画cities = df["城市"].unique()colors = ["red", "blue", "green", "orange", "purple"]for i, city in enumerate(cities):    city_data = df[df["城市"] == city]    plt.scatter(city_data["温度"], city_data["湿度"],                color=colors[i], label=city, alpha=0.5, s=20)plt.xlabel("温度 (C)", fontsize=14)plt.ylabel("湿度 (%)", fontsize=14)plt.title("温度 vs 湿度 散点图", fontsize=16)plt.legend(fontsize=12)plt.tight_layout()plt.show()
```

```python
# 观察下雨天和不下雨天的湿度差异plt.figure(figsize=(10, 6))rainy = df[df["是否下雨"] == 1]not_rainy = df[df["是否下雨"] == 0]plt.hist(not_rainy["湿度"], bins=20, alpha=0.6, label="不下雨", color="skyblue")plt.hist(rainy["湿度"], bins=20, alpha=0.6, label="下雨", color="coral")plt.xlabel("湿度 (%)", fontsize=14)plt.ylabel("天数", fontsize=14)plt.title("下雨天 vs 不下雨天的湿度分布", fontsize=16)plt.legend(fontsize=12)plt.tight_layout()plt.show()
```

```python
# 各城市平均温度柱状图plt.figure(figsize=(10, 6))city_temp = df.groupby("城市")["温度"].mean().sort_values(ascending=False)bars = plt.bar(city_temp.index, city_temp.values, color="steelblue")# 在柱子上显示数值for bar, val in zip(bars, city_temp.values):    plt.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.3,             f"{val:.1f}", ha="center", fontsize=11)plt.xlabel("城市", fontsize=14)plt.ylabel("平均温度 (C)", fontsize=14)plt.title("各城市平均温度对比", fontsize=16)plt.tight_layout()plt.show()
```

```python
# 相关性热力图：看看哪些因素有关联plt.figure(figsize=(8, 6))# 选择数值列numeric_cols = ["温度", "湿度", "风速", "气压", "是否下雨", "降雨量"]corr = df[numeric_cols].corr()sns.heatmap(corr, annot=True, cmap="RdBu_r", center=0,            fmt=".2f", square=True, linewidths=0.5)plt.title("天气因素相关性热力图", fontsize=16)plt.tight_layout()plt.show()
```
