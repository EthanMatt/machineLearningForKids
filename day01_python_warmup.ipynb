## 2. 列表 (List)列表就像一串糖葫芦，可以按顺序存放多个数据。

```python
# 创建一个列表temperatures = [23.5, 25.1, 22.8, 26.3, 24.0]cities = ["北京", "上海", "广州", "成都", "哈尔滨"]# 访问列表中的元素（注意：从 0 开始数！）print(f"第一个城市: {cities[0]}")print(f"最后一个城市: {cities[-1]}")print(f"温度列表的长度: {len(temperatures)}")
```

```python
# 列表的常用操作cities.append("杭州")print(f"添加后: {cities}")temperatures.sort()print(f"温度排序后: {temperatures}")# 列表推导式（快速生成列表）squares = [x**2 for x in range(1, 6)]print(f"1-5 的平方: {squares}")
```

## 3. 字典 (Dictionary)字典就像一本通讯录，通过 **键(key)** 查找 **值(value)**。

```python
# 创建一个字典weather = {    "城市": "北京",    "温度": 25,    "湿度": 60,    "天气": "晴"}# 访问字典print(f"城市: {weather[chr(224)+chr(227)+chr(230)+chr(236)]}")print(f"天气: {weather[chr(236)+chr(236)+chr(230)+chr(236)]}")# 添加新的键值对weather["风速"] = 3.5print(weather)
```

## 4. 循环 (Loop)循环可以让我们 **重复做某件事**，不用一遍一遍写同样的代码。

```python
# for 循环：遍历列表cities = ["北京", "上海", "广州", "成都", "哈尔滨"]for city in cities:    print(f"欢迎来到 {city}！")
```

```python
# while 循环：条件满足就一直做count = 0while count < 5:    print(f"这是第 {count+1} 次")    count += 1
```

## 5. 函数 (Function)函数就像一个 **魔法盒子**：你放进去一些东西，它帮你处理好再返回结果。

```python
# 定义一个简单的函数def greet(name):    """打招呼"""    return f"你好，{name}！"# 使用函数message = greet("小红")print(message)
```

```python
# 定义一个温度转换函数def celsius_to_fahrenheit(celsius):    """摄氏度转华氏度"""    return celsius * 9/5 + 32print(f"25C = {celsius_to_fahrenheit(25):.1f}F")print(f"0C = {celsius_to_fahrenheit(0):.1f}F")
```

---## 实战：读取天气数据现在让我们用 **pandas** 来读取真正的天气数据！

```python
# 导入 pandas 库import pandas as pd# 读取 CSV 文件df = pd.read_csv("data/weather_sample.csv", encoding="utf-8-sig")# 看前 5 行print("天气数据预览:")df.head()
```

```python
# 看数据的基本信息print(f"数据形状: {df.shape}")print(f"列名: {list(df.columns)}")print(f"数据类型:")print(df.dtypes)print(f"基本统计:")df.describe()
```

```python
# 找出每个城市的平均温度city_temp = df.groupby("城市")["温度"].mean()print("各城市平均温度:")print(city_temp.round(1))# 找出最热的城市hottest_city = city_temp.idxmax()hottest_temp = city_temp.max()print(f"最热的城市: {hottest_city}，平均温度 {hottest_temp:.1f}C")
```

```python
# 找出每个城市的平均湿度city_humidity = df.groupby("城市")["湿度"].mean()print("各城市平均湿度:")print(city_humidity.round(1))wettest_city = city_humidity.idxmax()print(f"最潮湿的城市: {wettest_city}")
```
